```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
def binaryStep(x):
    return np.heaviside(x,1)

x=np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation function :binaryStep')
plt.show()

def linear(x):
    return x
```

```python
x=np.linspace(-10, 10)
plt.plot(x, linear(x))
#plt.axis('tight')
plt.title('Activation function :linear')
plt.show()

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation function :sigmoid')
plt.show()

def tanh(x):
    return np.tanh(x)
```

```python
x=np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation function :tanh')
plt.show()

def RELU(x):
    x1= []
    for i in x:
        if i < 0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
```

```python
x=np.linspace(-10, 10)
plt.plot(x, RELU(x))
#plt.axis('tight')
plt.title('Activation function :RELU')
plt.show()

def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0)
```

```python
x=np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation function :softmax')
plt.show()
```
